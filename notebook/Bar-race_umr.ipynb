# Data visualization: Bar chart race

Library: https://www.dexplo.org/bar_chart_race/

Data source: https://www.bps.go.id/linkTableDinamis/view/id/917


```python
import bar_chart_race as bcr
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import seaborn as sns

%matplotlib inline 

df=pd.read_excel('./../data/bps-file-umr.xlsx')
print(df.shape)

tahun=df.columns
tahun=tahun.drop('Provinsi')
tahun
for x in (tahun):
    df[x]=pd.to_numeric(df[x],errors='coerce')

    
# df[tahun]=df[tahun].fillna(method='bfill',axis=1)    
df[tahun]=df[tahun].fillna(method='ffill',axis=1)    
df[tahun]=df[tahun].fillna(0,axis=1)    
df=df.transpose()
df=df.rename(columns=df.iloc[0])
df=df[1:len(df)]

df.index=pd.to_datetime(df.index, format='%Y')
df.index.name = 'Tahun'
for x in (df.columns):
    df[x]=pd.to_numeric(df[x],errors='coerce')

# tahun = list(map(str, range(1997, 2017)))
# df
```

```python
brc_filename="umr_1997_2016.mp4"
brc_title='Upah Minimum Regional/Provinsi (UMR/UMP) dalam Rupiah\n1997 - 2016'
bcr_html = bcr.bar_chart_race(
    df=df,
    filename=None,
    n_bars=7,
    filter_column_colors=True,
    period_fmt='Year: %Y',
    title=brc_title,
    title_size='',
    fixed_max=True,
)
bcr_html
```
